# Distribution_Plots

```python
attr_counts=df['Attrition'].value_counts()
print(attr_counts)

plt.figure(figsize=(6,4))
plt.bar(attr_counts.index, attr_counts.values)
plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition Status")
plt.ylabel("Number of Employees")

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["Age"], bins=20, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["Monthly Income"], bins=20, edgecolor="black")
plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(7,2))
plt.boxplot(df["Monthly Income"], vert=False)
plt.title("Monthly Income Boxplot")
plt.xlabel("Monthly Income")

plt.show()
```

```python

corr = df.drop(columns=["Employee ID"]).corr(numeric_only=True)
plt.figure(figsize=(10,8))
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")

plt.show()
```

# EDA Summary

## Key Findings

- The dataset contains **74,610 employee records** and **24 features**.
- Initial exploration was performed using `head()`, `tail()`, `shape`, `info()`, and `describe()` to understand the dataset structure.
- Two columns contain missing values:
  - **Distance from Home:** 1,912 missing values
  - **Company Tenure (In Months):** 2,413 missing values
- The dataset contains **112 duplicate records**, which will be removed during the preprocessing stage.
- The dataset includes both **numerical** and **categorical** features.
- The target variable **Attrition** consists of two classes:
  - **Stayed:** 39,191 employees
  - **Left:** 35,419 employees
- The target classes are relatively balanced, making the dataset suitable for classification.
- Distribution analysis was performed for **Age** and **Monthly Income** to understand the spread of numerical features.
- The **Monthly Income** boxplot indicates the presence of several outliers, including one extreme outlier.
- Correlation analysis shows that most numerical features have weak correlations. However, **Age**, **Years at Company**, and **Company Tenure (In Months)** exhibit moderate positive relationships.

## Conclusion

The exploratory data analysis provided a clear understanding of the dataset structure, feature types, data quality, and relationships among variables. Missing values, duplicate records, and outliers were identified during this phase. These issues will be addressed in the preprocessing stage before feature engineering and model training.
